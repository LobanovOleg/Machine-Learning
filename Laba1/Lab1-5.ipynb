````python
import pandas as pd
````

````python
df = pd.read_csv('data.csv')
````

1. Узнайте сколько было на борту человек

````python
x = df.shape[0]
print(x)
````

2. Какой медианный и средний возраста пассажиров

````python
x = df['Age'].mean()
y = df['Age'].median()
print("Mean:", x)
print("Median:", y)
````

3. Посчитайте процент выживаемости детей(до 16 лет) и взрослых

````python
children = df[df['Age'] < 16]
adults = df[df['Age'] >= 16]
if len(children) > 0:
    survival_children = children['Survived'].mean() * 100
else:
    survival_children = 0

if len(adults) > 0:
    survival_adults = adults['Survived'].mean() * 100
else:
    survival_adults = 0

print("Children: {:.2f}%".format(survival_children))
print("Adults: {:.2f}%".format(survival_adults))
````

4. Верна ли гипотеза, что женщины и дети сажались в шлюпки первыми и выживали больше?

````python
women = df[df['Sex'] == 'female']
men = df[df['Sex'] == 'male']
children = df[df['Age'] < 16]

survival_women = women['Survived'].mean() * 100
survival_men = men['Survived'].mean() * 100
if len(children) > 0:
    survival_children = children['Survived'].mean() * 100
else:
    survival_children = 0
print("Women: {:.2f}%".format(survival_women))
print("Men: {:.2f}%".format(survival_men))
print("Children: {:.2f}%".format(survival_children))
````

5. Зависит ли выживаемость от класса обслуживания?

````python
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("Survival by class:", survival_by_class)
````

6. Посчитайте средний возраст умерших женщин и мужчин

````python
women = df[(df['Survived'] == 0) & (df['Sex'] == 'female')]
men = df[(df['Survived'] == 0) & (df['Sex'] == 'male')]

if len(women) > 0:
    rate_W = women['Age'].mean()
else:
    rate_W = 0
if len(men) > 0:
    rate_M = men['Age'].mean()
else:
    rate_M = 0

print("Women: {:.2f}".format(rate_W))
print("Men: {:.2f}".format(rate_M))
````

7. Различается ли процент выживаемости пассажиров с хотя бы одним родственником на борту и одиночек?

````python
pas_with = df[(df['SibSp'] != 0) | (df['Parch'] != 0)]
pas_without = df[(df['SibSp'] == 0) & (df['Parch'] == 0)]

if len(pas_with) > 0:
    survived_with = pas_with['Survived'].mean() * 100
else:
    survived_with = 0

if len(pas_without) > 0:
    survived_without = pas_without['Survived'].mean() * 100
else:
    survived_without = 0

print("Survived with: {:.2f}%".format(survived_with))
print("Survived without: {:.2f}%".format(survived_without))
````

8. Различается ли средняя стоимость билета у умерших и выживших пассажиров?

````python
Survived = df[df['Survived'] == 1]
Dead = df[df['Survived'] == 0]

Fare_surv = Survived['Fare'].mean()
Fare_dead = Dead['Fare'].mean()

print("For survived: {:.2f}".format(Fare_surv))
print("For dead: {:.2f}".format(Fare_dead))
````

9. Выведите максимальный и минимальный возраст выживших пассажиров во 2 классе обслуживания

````python
Survived = df[(df['Survived'] == 1) & (df['Pclass'] == 2)]

max_age = Survived['Age'].max()
min_age = Survived['Age'].min()

print("Max: ", max_age)
print("Min: ", min_age)
````

10. Подтвердите или опровергните гипотезу: молодым мужчинам(от 18 лет) выжить легче, чем более взрослым

````python
men = df[df['Sex'] == 'male']
young = men[(men['Age'] >= 18) & (men['Age'] <= 35)]
old = men[men['Age'] > 35]
if len(young) > 0:
    survived_young = young['Survived'].mean() * 100
else:
    survived_young = 0
if len(old) > 0:
    survived_old = old['Survived'].mean() * 100
else:
    survived_old = 0

print("Young: {:.2f}%".format(survived_young))
print("Old: {:.2f}%".format(survived_old))
````

5. Визуализируйте распределение долей выживших среди мужчин, женщин и детей(до 16 лет)

````python
df['Category'] = pd.cut(df['Age'], bins=[0, 16], labels=['Дети'], right=False)
df['Category'] = df['Category'].cat.add_categories(['Мужчины', 'Женщины'])

df.loc[df['Sex'] == 'male', 'Category'] = 'Мужчины'
df.loc[df['Sex'] == 'female', 'Category'] = 'Женщины'

survived = df.groupby('Category', observed=False)['Survived'].mean()

plt.figure(figsize=(10, 6))
survived.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Доля выживших среди мужчин, женщин и детей')
plt.xlabel('Категория')
plt.ylabel('Доля выживших')
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
````

6. Сравните плотности распределения возростов выживших и умерших пассажиров, сделайте выводы

````python
df = df.dropna(subset=['Age'])
age_survived = df[df['Survived'] == 1]['Age']
age_dead = df[df['Survived'] == 0]['Age']
age_range = np.linspace(0, 80, 30)

survived, _ = np.histogram(age_survived, bins=age_range, density=True)

dead, _ = np.histogram(age_dead, bins=age_range, density=True)

plt.figure(figsize=(10, 6))
plt.fill_between(age_range[:-1], survived, color='blue', alpha=0.5, label='Выжившие')
plt.fill_between(age_range[:-1], dead, color='red', alpha=0.5, label='Умершие')

plt.title('Сравнение плотностей распределения возрастов выживших и умерших пассажиров')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend()
plt.grid(True)
plt.xlim(0, 80)
plt.show()
````
